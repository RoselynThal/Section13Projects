- Data generated by bikesharing systems--often automatically recorded--can be used to get a sense of mobility in cities, improve traffic, and detect important citywide events.

- **Capital Bikeshare serving the DC/VA area, along with city administrators, want to know whether traffic generated by the volume of rental bikes warrants strategies to improve traffic flow and biker safety in parts of the city.**

- *Has the volume of rental bikes increased annually from 2011 to 2012?*

- *Has the volume of rental bikes varied by weekday? Workday or offday?*

## Has the volume of rental bikes increased annually from 2011 to 2012?

```python
#Set up environment
import math
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
#https://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset


#Read in dataset
data = pd.read_csv('https://raw.githubusercontent.com/RoselynThal/Programming-Fundamentals/master/bikeshare_day.csv')

data.info()
data.head(10)
```

```python
#Split up dataset

data11 = data[(data['yr'] == 0)]
data12 = data[data.yr == 1]

data11.info()
data12.info()
```

### The median count of rental bikes increased from 3740 bikes in 2011 and 5927 bikes in 2012.

```python
#plt.barh(sales_sorted['country'], sales_sorted['sales'])

year_sorted = data.groupby(['yr'])['cnt'].aggregate(np.median).reset_index().sort_values('cnt')

fig, ax = plt.subplots(1, 1, figsize=(15, 5))
year_sorted.plot(kind='barh', x= 'yr', y = 'cnt', ax=ax, color='navy')

fmt = '{x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.xaxis.set_major_formatter(tick) 


plt.show()
```

- Is there a significant difference between the count of total active rental bikes in the system from 2011 to 2012? 


### Yes

```python
import seaborn as sns

# Set figure size
sns.set(rc={'figure.figsize':(8,6)})

g = sns.pointplot(data=[data11['cnt'],
                        data12['cnt']], join=False)
                        
# Adding a semicolon means output won't render in 
g.set(xticklabels = ['2011', '2012']);
```

## Has the volume of rental bikes varied by workday or offday?

```python
#Split up dataset

workday_data = data[(data['workingday'] == 1)]
offday_data = data[data.workingday == 0]

workday_data.info()
offday_data.info()
```

### There is no significant difference in total rental bikes based on whether it's a workday or an offday (a holiday or weekend).

- The mean count of rental bikes on workdays was 4584.82 and 4330.17 on offdays

```python
#plt.barh(work_sorted['workingday'], work_sorted['cnt'])

work_sorted = data.groupby(['workingday'])['cnt'].aggregate(np.average).reset_index().sort_values('cnt')

fig, ax = plt.subplots(1, 1, figsize=(15, 5))
work_sorted.plot(kind='barh', x= 'workingday', y = 'cnt', ax=ax, color='green')

fmt = '{x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.xaxis.set_major_formatter(tick) 


plt.show()
```

```python
#Confidence Interval

def get_95_ci(array_1, array_2):
    sample_1_n = array_1.shape[0]
    sample_2_n = array_2.shape[0]
    sample_1_mean = array_1.mean()
    sample_2_mean = array_2.mean()
    sample_1_var = array_1.var()
    sample_2_var = array_2.var()
    mean_difference = sample_2_mean - sample_1_mean
    std_err_difference = math.sqrt((sample_1_var/sample_1_n)+(sample_2_var/sample_2_n))
    margin_of_error = 1.96 * std_err_difference
    ci_lower = mean_difference - margin_of_error
    ci_upper = mean_difference + margin_of_error
    return("The difference in means at the 95% confidence interval (two-tail) is between "+str(ci_lower)+" and "+str(ci_upper)+".")

print(get_95_ci(workday_data.cnt, offday_data.cnt))
```

## Has the volume of rental bikes varied by weekday?

```python
#Create individual datasets for each day

Sunday = data[data.weekday == 0]
Monday = data[data.weekday == 1]
Tuesday = data[data.weekday == 2]
Wednesday = data[data.weekday == 3]
Thursday = data[data.weekday == 4]
Friday = data[data.weekday == 5]
Saturday = data[data.weekday == 6]

#Get descriptive statistics for each dataset

print(stats.describe(Sunday.cnt))
print(stats.describe(Monday.cnt))
print(stats.describe(Tuesday.cnt))
print(stats.describe(Wednesday.cnt))
print(stats.describe(Thursday.cnt))
print(stats.describe(Friday.cnt))
print(stats.describe(Saturday.cnt))

```

### There is no signficant difference in the number of active rental bikes based on weekday.

- The mean count of rental bikes overall was 4504.35 overall.

```python
#Create a boxplot to compare frequencies of cnts by weekday

data.boxplot('cnt', by = 'weekday', figsize=(10,6))
plt.show()
```
